```python
# Customer Support Chatbot - Google Colab Version (Fixed)
# Install required packages
!pip install tensorflow nltk scikit-learn flask streamlit

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
import re
import json
import pickle
from collections import Counter
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Download NLTK data (including the missing punkt_tab)
print("Downloading NLTK data...")
nltk.download('punkt')
nltk.download('punkt_tab')  # This fixes the LookupError
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Embedding, LSTM
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical

# Set random seeds
np.random.seed(42)
tf.random.set_seed(42)

print("Customer Support Chatbot Development")
print("=" * 50)

# ============================================================================
# DATA GENERATION AND PREPROCESSING
# ============================================================================

def generate_customer_support_data():
    """
    Generate synthetic customer support data with intents and responses
    """
    print("Generating customer support dataset...")

    # Define intents and sample queries
    intents_data = {
        "greeting": {
            "patterns": [
                "hello", "hi", "hey", "good morning", "good afternoon", "good evening",
                "greetings", "what's up", "howdy", "hi there", "hello there"
            ],
            "responses": [
                "Hello! How can I help you today?",
                "Hi there! What can I assist you with?",
                "Greetings! How may I help you?",
                "Hello! I'm here to help. What do you need?",
                "Hi! What can I do for you today?"
            ]
        },
        "goodbye": {
            "patterns": [
                "bye", "goodbye", "see you later", "farewell", "talk to you later",
                "catch you later", "until next time", "bye bye", "see ya", "later"
            ],
            "responses": [
                "Goodbye! Have a great day!",
                "See you later! Feel free to reach out anytime.",
                "Farewell! Thank you for contacting us.",
                "Bye! Hope I was able to help you today.",
                "Take care! Don't hesitate to contact us again."
            ]
        },
        "account_issues": {
            "patterns": [
                "can't login", "forgot password", "account locked", "login problem",
                "password reset", "account access", "login issues", "can't access account",
                "forgot username", "account recovery", "login error", "authentication problem"
            ],
            "responses": [
                "I can help you with account issues. Please try resetting your password using the 'Forgot Password' link.",
                "For account access problems, please visit our password reset page or contact our technical support.",
                "If you're having login issues, please check your email for password reset instructions.",
                "Account locked? Please wait 15 minutes and try again, or contact support for immediate assistance.",
                "For account recovery, please provide your registered email address to our support team."
            ]
        },
        "billing": {
            "patterns": [
                "billing question", "payment issue", "invoice", "charge", "refund",
                "subscription", "payment method", "billing address", "payment failed",
                "credit card", "billing cycle", "payment history", "billing support"
            ],
            "responses": [
                "For billing inquiries, please check your account dashboard or contact our billing department.",
                "Payment issues can be resolved by updating your payment method in account settings.",
                "For refund requests, please contact our billing team with your order number.",
                "Billing questions? Please review your invoice in the billing section of your account.",
                "Payment problems? Please verify your payment method and billing address."
            ]
        },
        "technical_support": {
            "patterns": [
                "technical issue", "bug", "error", "not working", "broken",
                "technical problem", "system error", "application error", "website down",
                "server error", "loading issue", "performance problem", "technical help"
            ],
            "responses": [
                "I'm sorry you're experiencing technical issues. Please try clearing your browser cache and cookies.",
                "For technical problems, please check our status page or contact our technical support team.",
                "Technical issues can often be resolved by refreshing the page or restarting the application.",
                "If you're experiencing bugs, please report them to our technical team with detailed steps to reproduce.",
                "For system errors, please try again later or contact our technical support for immediate assistance."
            ]
        },
        "product_info": {
            "patterns": [
                "product information", "features", "specifications", "product details",
                "what does it do", "how does it work", "product benefits", "capabilities",
                "product comparison", "product specs", "functionality", "product overview"
            ],
            "responses": [
                "You can find detailed product information on our website's product pages.",
                "For specific product features and specifications, please visit our documentation section.",
                "Product details are available in our catalog. What specific information are you looking for?",
                "Our products offer various features. Please check our website for comprehensive details.",
                "For product comparisons and specifications, please visit our products section."
            ]
        },
        "shipping": {
            "patterns": [
                "shipping", "delivery", "tracking", "order status", "when will it arrive",
                "shipping cost", "delivery time", "tracking number", "order tracking",
                "shipping address", "delivery status", "shipping options", "express delivery"
            ],
            "responses": [
                "You can track your order using the tracking number sent to your email.",
                "Shipping information is available in your account under 'Order History'.",
                "For delivery questions, please check your order status or contact our shipping department.",
                "Standard shipping takes 3-5 business days. Express options are available at checkout.",
                "To update your shipping address, please contact us before the item ships."
            ]
        },
        "returns": {
            "patterns": [
                "return", "exchange", "return policy", "refund policy", "send back",
                "return item", "exchange product", "return process", "how to return",
                "return shipping", "return label", "return authorization", "return request"
            ],
            "responses": [
                "Our return policy allows returns within 30 days of purchase. Please check our returns page.",
                "To return an item, please initiate a return request in your account or contact customer service.",
                "Return shipping labels can be generated from your account's order history section.",
                "For exchanges, please return the original item and place a new order for the replacement.",
                "Return authorization numbers can be obtained by contacting our customer service team."
            ]
        },
        "hours": {
            "patterns": [
                "business hours", "operating hours", "when are you open", "store hours",
                "customer service hours", "support hours", "office hours", "availability",
                "when can I call", "contact hours", "working hours", "service time"
            ],
            "responses": [
                "Our customer service is available Monday-Friday 9AM-6PM EST.",
                "Support hours are 9:00 AM to 6:00 PM, Monday through Friday.",
                "We're open for customer service Monday to Friday, 9 AM to 6 PM Eastern Time.",
                "Our business hours are 9:00 AM - 6:00 PM EST, Monday through Friday.",
                "Customer support is available weekdays from 9 AM to 6 PM Eastern Time."
            ]
        },
        "contact": {
            "patterns": [
                "contact", "phone number", "email", "address", "how to reach",
                "customer service number", "support email", "contact information",
                "get in touch", "reach out", "contact details", "customer service contact"
            ],
            "responses": [
                "You can contact us at support@company.com or call 1-800-123-4567.",
                "Our customer service email is support@company.com and phone is 1-800-123-4567.",
                "Contact information: Email: support@company.com, Phone: 1-800-123-4567",
                "Reach us at support@company.com or call our customer service at 1-800-123-4567.",
                "For immediate assistance, email support@company.com or call 1-800-123-4567."
            ]
        }
    }

    # Generate training data
    training_data = []
    for intent, data in intents_data.items():
        for pattern in data["patterns"]:
            training_data.append({
                "text": pattern,
                "intent": intent,
                "response": np.random.choice(data["responses"])
            })

    # Add variations and more realistic queries
    variations = []
    for item in training_data:
        text = item["text"]
        intent = item["intent"]
        response = item["response"]

        # Add question variations
        variations.extend([
            {"text": f"I need help with {text}", "intent": intent, "response": response},
            {"text": f"Can you help me with {text}?", "intent": intent, "response": response},
            {"text": f"I have a problem with {text}", "intent": intent, "response": response},
            {"text": f"What about {text}?", "intent": intent, "response": response},
            {"text": f"Tell me about {text}", "intent": intent, "response": response}
        ])

    training_data.extend(variations)

    # Convert to DataFrame
    df = pd.DataFrame(training_data)

    print(f"Generated {len(df)} training samples")
    print(f"Number of intents: {df['intent'].nunique()}")

    return df, intents_data

# Generate the dataset
training_df, intents_dict = generate_customer_support_data()

# Display dataset information
print(f"\nDataset Overview:")
print(f"Total samples: {len(training_df)}")
print(f"Unique intents: {training_df['intent'].nunique()}")
print(f"\nIntent distribution:")
print(training_df['intent'].value_counts())

# ============================================================================
# TEXT PREPROCESSING (FIXED VERSION)
# ============================================================================

class TextPreprocessor:
    def __init__(self):
        self.lemmatizer = WordNetLemmatizer()
        self.stop_words = set(stopwords.words('english'))

    def preprocess_text(self, text):
        """
        Preprocess text for NLP tasks with error handling
        """
        try:
            # Convert to lowercase
            text = text.lower()

            # Remove special characters and digits
            text = re.sub(r'[^a-zA-Z\s]', '', text)

            # Tokenize with error handling
            try:
                tokens = word_tokenize(text)
            except LookupError:
                # Fallback to simple split if tokenizer fails
                tokens = text.split()

            # Remove stopwords and lemmatize
            processed_tokens = []
            for token in tokens:
                if token not in self.stop_words and len(token) > 2:
                    try:
                        lemmatized = self.lemmatizer.lemmatize(token)
                        processed_tokens.append(lemmatized)
                    except:
                        # If lemmatization fails, use original token
                        processed_tokens.append(token)

            return ' '.join(processed_tokens)

        except Exception as e:
            print(f"Error preprocessing text: {e}")
            # Return cleaned text without advanced preprocessing
            return re.sub(r'[^a-zA-Z\s]', '', text.lower())

    def preprocess_dataset(self, df):
        """
        Preprocess entire dataset with error handling
        """
        print("Preprocessing text data...")
        df['processed_text'] = df['text'].apply(self.preprocess_text)
        return df

# Initialize preprocessor and process data
preprocessor = TextPreprocessor()
training_df = preprocessor.preprocess_dataset(training_df)

print(f"\nSample preprocessed data:")
for i in range(3):
    print(f"Original: {training_df.iloc[i]['text']}")
    print(f"Processed: {training_df.iloc[i]['processed_text']}")
    print(f"Intent: {training_df.iloc[i]['intent']}")
    print("-" * 40)

# ============================================================================
# FEATURE EXTRACTION AND MODEL TRAINING
# ============================================================================

print("\nPreparing data for machine learning...")

# Prepare features and labels
X = training_df['processed_text']
y = training_df['intent']

# Encode labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print(f"Training samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer(max_features=1000, ngram_range=(1, 2))
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# ============================================================================
# TRADITIONAL ML MODELS
# ============================================================================

print("\nTraining traditional ML models...")

# Initialize models
ml_models = {
    'Naive Bayes': MultinomialNB(),
    'SVM': SVC(kernel='linear', probability=True),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42)
}

# Train and evaluate models
model_results = {}

for name, model in ml_models.items():
    print(f"\nTraining {name}...")

    # Train model
    model.fit(X_train_tfidf, y_train)

    # Make predictions
    y_pred = model.predict(X_test_tfidf)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)

    # Store results
    model_results[name] = {
        'model': model,
        'accuracy': accuracy,
        'predictions': y_pred
    }

    print(f"{name} Accuracy: {accuracy:.4f}")

# Find best traditional ML model
best_ml_model_name = max(model_results.keys(), key=lambda x: model_results[x]['accuracy'])
best_ml_model = model_results[best_ml_model_name]['model']

print(f"\nBest traditional ML model: {best_ml_model_name} ({model_results[best_ml_model_name]['accuracy']:.4f})")

# ============================================================================
# SIMPLIFIED DEEP LEARNING MODEL
# ============================================================================

print("\nTraining simplified neural network model...")

# Prepare data for neural network
# Use TF-IDF features instead of sequences to avoid tokenization issues
X_train_dense = X_train_tfidf.toarray()
X_test_dense = X_test_tfidf.toarray()

# Convert labels to categorical
num_classes = len(label_encoder.classes_)
y_train_cat = to_categorical(y_train, num_classes)
y_test_cat = to_categorical(y_test, num_classes)

# Build simple neural network model
nn_model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_dense.shape[1],)),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')
])

nn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("Neural Network Model Architecture:")
nn_model.summary()

# Train neural network model
print("\nTraining neural network model...")
history = nn_model.fit(
    X_train_dense, y_train_cat,
    epochs=20,
    batch_size=32,
    validation_data=(X_test_dense, y_test_cat),
    verbose=1
)

# Evaluate neural network model
nn_loss, nn_accuracy = nn_model.evaluate(X_test_dense, y_test_cat, verbose=0)
print(f"\nNeural Network Model Accuracy: {nn_accuracy:.4f}")

# ============================================================================
# CHATBOT CLASS (SIMPLIFIED)
# ============================================================================

class CustomerSupportChatbot:
    def __init__(self, ml_model, nn_model, vectorizer, label_encoder,
                 preprocessor, intents_dict):
        self.ml_model = ml_model
        self.nn_model = nn_model
        self.vectorizer = vectorizer
        self.label_encoder = label_encoder
        self.preprocessor = preprocessor
        self.intents_dict = intents_dict
        self.conversation_history = []

    def predict_intent_ml(self, text):
        """Predict intent using traditional ML model"""
        processed_text = self.preprocessor.preprocess_text(text)
        text_vector = self.vectorizer.transform([processed_text])

        # Get prediction and confidence
        prediction = self.ml_model.predict(text_vector)[0]
        confidence = max(self.ml_model.predict_proba(text_vector)[0])

        intent = self.label_encoder.inverse_transform([prediction])[0]
        return intent, confidence

    def predict_intent_nn(self, text):
        """Predict intent using neural network model"""
        processed_text = self.preprocessor.preprocess_text(text)
        text_vector = self.vectorizer.transform([processed_text]).toarray()

        # Get prediction and confidence
        prediction = self.nn_model.predict(text_vector, verbose=0)[0]
        confidence = max(prediction)
        intent_index = np.argmax(prediction)

        intent = self.label_encoder.inverse_transform([intent_index])[0]
        return intent, confidence

    def get_response(self, text, use_nn=True, confidence_threshold=0.5):
        """Get chatbot response"""
        # Predict intent
        if use_nn:
            intent, confidence = self.predict_intent_nn(text)
        else:
            intent, confidence = self.predict_intent_ml(text)

        # Log conversation
        self.conversation_history.append({
            'user_input': text,
            'predicted_intent': intent,
            'confidence': confidence,
            'timestamp': datetime.now()
        })

        # Generate response
        if confidence < confidence_threshold:
            response = "I'm not sure I understand. Could you please rephrase your question or contact our customer service team?"
        else:
            responses = self.intents_dict[intent]['responses']
            response = np.random.choice(responses)

        return response, intent, confidence

    def get_conversation_stats(self):
        """Get conversation statistics"""
        if not self.conversation_history:
            return "No conversations yet."

        intents = [conv['predicted_intent'] for conv in self.conversation_history]
        intent_counts = Counter(intents)
        avg_confidence = np.mean([conv['confidence'] for conv in self.conversation_history])

        return {
            'total_conversations': len(self.conversation_history),
            'unique_intents': len(set(intents)),
            'most_common_intents': intent_counts.most_common(5),
            'average_confidence': avg_confidence
        }

# Initialize chatbot
chatbot = CustomerSupportChatbot(
    ml_model=best_ml_model,
    nn_model=nn_model,
    vectorizer=tfidf_vectorizer,
    label_encoder=label_encoder,
    preprocessor=preprocessor,
    intents_dict=intents_dict
)

print(f"\nChatbot initialized successfully!")

# ============================================================================
# MODEL EVALUATION AND COMPARISON
# ============================================================================

print("\nEvaluating and comparing models...")

# Create comparison plots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# 1. Model accuracy comparison
model_names = list(model_results.keys()) + ['Neural Network']
accuracies = [model_results[name]['accuracy'] for name in model_results.keys()] + [nn_accuracy]

axes[0, 0].bar(model_names, accuracies, alpha=0.7)
axes[0, 0].set_title('Model Accuracy Comparison')
axes[0, 0].set_ylabel('Accuracy')
axes[0, 0].tick_params(axis='x', rotation=45)

# 2. Intent distribution
intent_counts = training_df['intent'].value_counts()
axes[0, 1].pie(intent_counts.values, labels=intent_counts.index, autopct='%1.1f%%')
axes[0, 1].set_title('Intent Distribution in Training Data')

# 3. Neural network training history
axes[1, 0].plot(history.history['accuracy'], label='Training Accuracy')
axes[1, 0].plot(history.history['val_accuracy'], label='Validation Accuracy')
axes[1, 0].set_title('Neural Network Training History')
axes[1, 0].set_xlabel('Epoch')
axes[1, 0].set_ylabel('Accuracy')
axes[1, 0].legend()

# 4. Confusion matrix for best model
best_model_pred = model_results[best_ml_model_name]['predictions']
cm = confusion_matrix(y_test, best_model_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[1, 1])
axes[1, 1].set_title(f'Confusion Matrix - {best_ml_model_name}')
axes[1, 1].set_xlabel('Predicted')
axes[1, 1].set_ylabel('Actual')

plt.tight_layout()
plt.show()

# Print detailed classification report
print(f"\nDetailed Classification Report for {best_ml_model_name}:")
print(classification_report(y_test, best_model_pred,
                          target_names=label_encoder.classes_))

# ============================================================================
# CHATBOT TESTING
# ============================================================================

# Test chatbot with sample queries
print("\nTesting chatbot with sample queries...")

sample_queries = [
    "Hello, I need help",
    "I can't login to my account",
    "What are your business hours?",
    "I want to return a product",
    "How much does shipping cost?",
    "I have a billing question",
    "The website is not working",
    "Thank you, goodbye"
]

print("\nSample Conversations:")
print("-" * 50)

for query in sample_queries:
    response, intent, confidence = chatbot.get_response(query, use_nn=True)
    print(f"User: {query}")
    print(f"Bot: {response}")
    print(f"[Intent: {intent}, Confidence: {confidence:.3f}]")
    print("-" * 50)

# ============================================================================
# ANALYTICS AND INSIGHTS
# ============================================================================

print("\nGenerating analytics and insights...")

# Simulate some conversation data for analytics
np.random.seed(42)
simulated_conversations = []

for _ in range(100):
    intent = np.random.choice(list(intents_dict.keys()))
    confidence = np.random.uniform(0.6, 0.95)
    timestamp = datetime.now() - pd.Timedelta(days=np.random.randint(0, 30))

    simulated_conversations.append({
        'intent': intent,
        'confidence': confidence,
        'timestamp': timestamp,
        'resolved': np.random.choice([True, False], p=[0.8, 0.2])
    })

analytics_df = pd.DataFrame(simulated_conversations)

# Create analytics dashboard
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# 1. Most common customer inquiries
intent_counts = analytics_df['intent'].value_counts()
axes[0, 0].bar(intent_counts.index, intent_counts.values)
axes[0, 0].set_title('Most Common Customer Inquiries')
axes[0, 0].set_ylabel('Number of Inquiries')
axes[0, 0].tick_params(axis='x', rotation=45)

# 2. Confidence distribution
axes[0, 1].hist(analytics_df['confidence'], bins=20, alpha=0.7, color='green')
axes[0, 1].set_title('Chatbot Confidence Distribution')
axes[0, 1].set_xlabel('Confidence Score')
axes[0, 1].set_ylabel('Frequency')

# 3. Resolution rate by intent
resolution_by_intent = analytics_df.groupby('intent')['resolved'].mean()
axes[1, 0].bar(resolution_by_intent.index, resolution_by_intent.values, color='orange')
axes[1, 0].set_title('Resolution Rate by Intent')
axes[1, 0].set_ylabel('Resolution Rate')
axes[1, 0].tick_params(axis='x', rotation=45)

# 4. Daily conversation volume
analytics_df['date'] = analytics_df['timestamp'].dt.date
daily_volume = analytics_df.groupby('date').size()
axes[1, 1].plot(daily_volume.index, daily_volume.values, marker='o')
axes[1, 1].set_title('Daily Conversation Volume')
axes[1, 1].set_xlabel('Date')
axes[1, 1].set_ylabel('Number of Conversations')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Print analytics insights
print(f"\nCustomer Support Analytics Insights:")
print(f"=" * 50)
print(f"Total Conversations Analyzed: {len(analytics_df)}")
print(f"Average Confidence Score: {analytics_df['confidence'].mean():.3f}")
print(f"Overall Resolution Rate: {analytics_df['resolved'].mean():.1%}")

print(f"\nTop 5 Customer Inquiries:")
for i, (intent, count) in enumerate(intent_counts.head().items(), 1):
    print(f"  {i}. {intent.replace('_', ' ').title()}: {count} inquiries")

print(f"\nModel Performance Summary:")
print(f"  • Best Traditional ML Model: {best_ml_model_name} ({model_results[best_ml_model_name]['accuracy']:.1%})")
print(f"  • Neural Network Model Accuracy: {nn_accuracy:.1%}")
print(f"  • Number of Intents Supported: {len(intents_dict)}")
print(f"  • Training Samples: {len(training_df)}")

print(f"\nSupported Customer Intents:")
for i, intent in enumerate(intents_dict.keys(), 1):
    print(f"  {i}. {intent.replace('_', ' ').title()}")

print(f"\nCustomer Support Chatbot Development Complete!")
print(f"All models trained successfully without NLTK errors!")

# Interactive demo function
def interactive_chatbot_demo():
    """
    Interactive chatbot demonstration
    """
    print("\n" + "="*60)
    print("CUSTOMER SUPPORT CHATBOT DEMO")
    print("="*60)
    print("Type 'quit' to exit the demo")
    print("Type 'stats' to see conversation statistics")
    print("-"*60)

    while True:
        user_input = input("\nYou: ").strip()

        if user_input.lower() == 'quit':
            print("Chatbot: Thank you for using our customer support! Goodbye!")
            break
        elif user_input.lower() == 'stats':
            stats = chatbot.get_conversation_stats()
            print(f"\nConversation Statistics:")
            for key, value in stats.items():
                print(f"  {key}: {value}")
            continue
        elif not user_input:
            continue

        # Get response from chatbot
        response, intent, confidence = chatbot.get_response(user_input, use_nn=True)

        print(f"Chatbot: {response}")
        print(f"[Intent: {intent}, Confidence: {confidence:.3f}]")

print(f"\nTo run interactive demo, call: interactive_chatbot_demo()")
```
