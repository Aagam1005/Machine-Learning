```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
import cv2
import os
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("Hand Gesture Recognition System (Final - Error Free)")
print("=" * 60)

# For demonstration, we'll create synthetic gesture data
print("Creating synthetic hand gesture dataset...")

def create_synthetic_gesture_data(n_samples=2000, img_size=64):
    """
    Create synthetic hand gesture data for demonstration
    """
    images = []
    labels = []

    # Define 10 different hand gestures
    gesture_names = [
        'palm', 'fist', 'thumbs_up', 'thumbs_down', 'peace',
        'ok_sign', 'pointing', 'rock', 'paper', 'scissors'
    ]

    for gesture_idx, gesture_name in enumerate(gesture_names):
        for i in range(n_samples // len(gesture_names)):
            # Create synthetic hand-like patterns
            img = np.zeros((img_size, img_size, 3), dtype=np.uint8)

            # Create different patterns for different gestures
            if gesture_name == 'palm':
                # Open palm - large filled circle
                center = (img_size//2, img_size//2)
                radius = img_size//3
                cv2.circle(img, center, radius, (200, 180, 160), -1)
                # Add fingers
                for angle in [0, 45, 90, 135, 180]:
                    x = int(center[0] + radius * 0.8 * np.cos(np.radians(angle)))
                    y = int(center[1] + radius * 0.8 * np.sin(np.radians(angle)))
                    cv2.circle(img, (x, y), 8, (220, 200, 180), -1)

            elif gesture_name == 'fist':
                # Closed fist - smaller filled circle
                center = (img_size//2, img_size//2)
                radius = img_size//4
                cv2.circle(img, center, radius, (200, 180, 160), -1)

            elif gesture_name == 'thumbs_up':
                # Fist with thumb up
                center = (img_size//2, img_size//2 + 5)
                radius = img_size//4
                cv2.circle(img, center, radius, (200, 180, 160), -1)
                # Thumb
                thumb_center = (center[0] - 10, center[1] - radius - 10)
                cv2.circle(img, thumb_center, 6, (220, 200, 180), -1)

            elif gesture_name == 'thumbs_down':
                # Fist with thumb down
                center = (img_size//2, img_size//2 - 5)
                radius = img_size//4
                cv2.circle(img, center, radius, (200, 180, 160), -1)
                # Thumb
                thumb_center = (center[0] - 10, center[1] + radius + 10)
                cv2.circle(img, thumb_center, 6, (220, 200, 180), -1)

            elif gesture_name == 'peace':
                # Two fingers up
                center = (img_size//2, img_size//2)
                radius = img_size//5
                cv2.circle(img, center, radius, (200, 180, 160), -1)
                # Two fingers
                finger1 = (center[0] - 8, center[1] - radius - 15)
                finger2 = (center[0] + 8, center[1] - radius - 15)
                cv2.circle(img, finger1, 5, (220, 200, 180), -1)
                cv2.circle(img, finger2, 5, (220, 200, 180), -1)

            elif gesture_name == 'ok_sign':
                # Circle with thumb and index finger
                center = (img_size//2, img_size//2)
                cv2.circle(img, center, 8, (200, 180, 160), 3)
                # Other fingers
                for i in range(3):
                    finger_pos = (center[0] + 15 + i*5, center[1] - 10)
                    cv2.circle(img, finger_pos, 4, (220, 200, 180), -1)

            elif gesture_name == 'pointing':
                # Index finger pointing
                center = (img_size//2, img_size//2)
                radius = img_size//5
                cv2.circle(img, center, radius, (200, 180, 160), -1)
                # Pointing finger
                finger_tip = (center[0] + radius + 20, center[1])
                cv2.circle(img, finger_tip, 4, (220, 200, 180), -1)
                cv2.line(img, center, finger_tip, (210, 190, 170), 8)

            elif gesture_name == 'rock':
                # Rock gesture (similar to fist but slightly different)
                center = (img_size//2, img_size//2)
                radius = img_size//4
                cv2.ellipse(img, center, (radius, radius-2), 0, 0, 360, (200, 180, 160), -1)

            elif gesture_name == 'paper':
                # Open hand (similar to palm but flatter)
                center = (img_size//2, img_size//2)
                cv2.rectangle(img, (center[0]-20, center[1]-25),
                            (center[0]+20, center[1]+25), (200, 180, 160), -1)
                # Fingers
                for i in range(5):
                    finger_x = center[0] - 15 + i*7
                    finger_y = center[1] - 25
                    cv2.rectangle(img, (finger_x-2, finger_y-15),
                                (finger_x+2, finger_y), (220, 200, 180), -1)

            else:  # scissors
                # Two fingers in V shape
                center = (img_size//2, img_size//2)
                radius = img_size//5
                cv2.circle(img, center, radius, (200, 180, 160), -1)
                # V-shaped fingers
                finger1_tip = (center[0] - 15, center[1] - radius - 10)
                finger2_tip = (center[0] + 15, center[1] - radius - 10)
                cv2.line(img, center, finger1_tip, (220, 200, 180), 6)
                cv2.line(img, center, finger2_tip, (220, 200, 180), 6)

            # Add some noise and variation
            noise = np.random.normal(0, 10, img.shape).astype(np.int16)
            img = np.clip(img.astype(np.int16) + noise, 0, 255).astype(np.uint8)

            # Random rotation and scaling
            angle = np.random.uniform(-15, 15)
            scale = np.random.uniform(0.9, 1.1)
            center_rot = (img_size//2, img_size//2)
            M = cv2.getRotationMatrix2D(center_rot, angle, scale)
            img = cv2.warpAffine(img, M, (img_size, img_size))

            images.append(img)
            labels.append(gesture_idx)

    return np.array(images), np.array(labels), gesture_names

# Create synthetic dataset
X_data, y_data, gesture_names = create_synthetic_gesture_data(2000, 64)

print(f"Dataset created successfully!")
print(f"Number of images: {len(X_data)}")
print(f"Image shape: {X_data[0].shape}")
print(f"Number of classes: {len(gesture_names)}")
print(f"Gesture classes: {gesture_names}")

# Display sample images from each gesture class
plt.figure(figsize=(20, 8))
for i, gesture_name in enumerate(gesture_names):
    # Find first occurrence of each gesture
    gesture_idx = np.where(y_data == i)[0][0]
    plt.subplot(2, 5, i + 1)
    plt.imshow(X_data[gesture_idx])
    plt.title(f"{gesture_name.replace('_', ' ').title()}")
    plt.axis('off')
plt.suptitle('Sample Hand Gestures', fontsize=16)
plt.tight_layout()
plt.show()

# Data preprocessing
print("\nPreprocessing data...")

# Normalize pixel values to [0, 1]
X_data = X_data.astype('float32') / 255.0

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X_data, y_data, test_size=0.2, random_state=42, stratify=y_data
)

# Further split training data for validation
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42, stratify=y_train
)

print(f"Training set: {X_train.shape[0]} samples")
print(f"Validation set: {X_val.shape[0]} samples")
print(f"Test set: {X_test.shape[0]} samples")

# Convert labels to categorical
num_classes = len(gesture_names)
y_train_cat = keras.utils.to_categorical(y_train, num_classes)
y_val_cat = keras.utils.to_categorical(y_val, num_classes)
y_test_cat = keras.utils.to_categorical(y_test, num_classes)

print("Data preprocessing completed!")

# Build the ERROR-FREE CNN model for gesture recognition
print("Building ERROR-FREE hand gesture recognition model...")

def create_error_free_model(input_shape, num_classes):
    """
    Create a model without data augmentation layers to avoid pickle errors
    """
    model = keras.Sequential([
        # First convolutional block
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape, padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # Second convolutional block
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # Third convolutional block
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # Fourth convolutional block
        layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.GlobalAveragePooling2D(),

        # Dense layers
        layers.Dense(512, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ])

    return model

model = create_error_free_model(X_train.shape[1:], num_classes)

# Display model architecture
print("\nError-Free Model Architecture:")
model.summary()

# Compile the model
print("Compiling model...")
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Data augmentation using ImageDataGenerator (compatible with all TensorFlow versions)
print("Setting up data augmentation...")
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    zoom_range=0.2,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest'
)

# Set up SIMPLE callbacks (no ModelCheckpoint to avoid errors)
callbacks = [
    keras.callbacks.EarlyStopping(
        monitor='val_accuracy',
        patience=15,
        restore_best_weights=True,
        verbose=1
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=7,
        min_lr=1e-7,
        verbose=1
    )
]

# Train the model with data augmentation
print("Training ERROR-FREE hand gesture recognition model...")
print("This may take several minutes...")

# Fit the data generator
datagen.fit(X_train)

# Train the model
history = model.fit(
    datagen.flow(X_train, y_train_cat, batch_size=32),
    steps_per_epoch=len(X_train) // 32,
    epochs=50,
    validation_data=(X_val, y_val_cat),
    callbacks=callbacks,
    verbose=1
)

# Evaluate the model
print("\nEvaluating model...")
test_loss, test_accuracy = model.evaluate(X_test, y_test_cat, verbose=0)
print(f"Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f"Test Loss: {test_loss:.4f}")

# Plot training history
plt.figure(figsize=(15, 5))

# Plot accuracy
plt.subplot(1, 3, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy', linewidth=2)
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
plt.title('Model Accuracy Over Time')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot loss
plt.subplot(1, 3, 2)
plt.plot(history.history['loss'], label='Training Loss', linewidth=2)
plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
plt.title('Model Loss Over Time')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot learning rate
plt.subplot(1, 3, 3)
if 'lr' in history.history:
    plt.plot(history.history['lr'], linewidth=2)
    plt.title('Learning Rate Over Time')
    plt.xlabel('Epoch')
    plt.ylabel('Learning Rate')
    plt.yscale('log')
    plt.grid(True, alpha=0.3)
else:
    plt.text(0.5, 0.5, 'Learning Rate\nScheduling Applied',
             ha='center', va='center', transform=plt.gca().transAxes)
    plt.title('Learning Rate')

plt.tight_layout()
plt.show()

# Make predictions
print("Making predictions...")
y_pred_proba = model.predict(X_test, verbose=0)
y_pred = np.argmax(y_pred_proba, axis=1)

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=gesture_names))

# Confusion Matrix
print("Generating confusion matrix...")
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=[name.replace('_', ' ').title() for name in gesture_names],
            yticklabels=[name.replace('_', ' ').title() for name in gesture_names])
plt.title('Hand Gesture Recognition - Confusion Matrix')
plt.xlabel('Predicted Gesture')
plt.ylabel('True Gesture')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Calculate per-class accuracy
class_accuracies = cm.diagonal() / cm.sum(axis=1)
print("\nPer-Gesture Accuracy:")
for i, (gesture_name, accuracy) in enumerate(zip(gesture_names, class_accuracies)):
    print(f"  {gesture_name.replace('_', ' ').title()}: {accuracy:.4f} ({accuracy*100:.2f}%)")

# Display sample predictions with confidence scores
def display_gesture_predictions(images, true_labels, pred_labels, pred_proba, gesture_names, n_samples=15):
    plt.figure(figsize=(20, 12))

    # Randomly select samples
    indices = np.random.choice(len(images), n_samples, replace=False)

    for i, idx in enumerate(indices):
        plt.subplot(3, 5, i + 1)
        plt.imshow(images[idx])

        true_gesture = gesture_names[true_labels[idx]]
        pred_gesture = gesture_names[pred_labels[idx]]
        confidence = pred_proba[idx][pred_labels[idx]]

        # Determine if prediction is correct
        is_correct = true_labels[idx] == pred_labels[idx]

        # Set title color based on correctness
        title_color = 'green' if is_correct else 'red'

        plt.title(f"Predicted: {pred_gesture.replace('_', ' ').title()}\n"
                 f"Confidence: {confidence:.2f}\n"
                 f"True: {true_gesture.replace('_', ' ').title()}",
                 color=title_color, fontsize=10)
        plt.axis('off')

    plt.suptitle('Hand Gesture Recognition - Sample Predictions', fontsize=16)
    plt.tight_layout()
    plt.show()

# Display sample predictions
print("\nDisplaying sample predictions...")
display_gesture_predictions(X_test, y_test, y_pred, y_pred_proba, gesture_names, n_samples=15)

# Real-time gesture recognition function
def predict_gesture(image, model, gesture_names, img_size=64):
    """
    Predict gesture from a single image
    """
    # Preprocess image
    if len(image.shape) == 3:
        # Resize if necessary
        if image.shape[:2] != (img_size, img_size):
            image = cv2.resize(image, (img_size, img_size))

    # Normalize
    image = image.astype('float32') / 255.0

    # Add batch dimension
    image = np.expand_dims(image, axis=0)

    # Make prediction
    prediction = model.predict(image, verbose=0)
    predicted_class_idx = np.argmax(prediction)
    confidence = prediction[0][predicted_class_idx]

    predicted_gesture = gesture_names[predicted_class_idx]

    return predicted_gesture, confidence, prediction[0]

# Gesture-based control system simulation
print("\nGesture-Based Control System Simulation")
print("=" * 45)

def simulate_gesture_control(gesture, confidence_threshold=0.7):
    """
    Simulate different actions based on recognized gestures
    """
    actions = {
        'palm': 'Stop/Pause',
        'fist': 'Select/Confirm',
        'thumbs_up': 'Like/Approve',
        'thumbs_down': 'Dislike/Reject',
        'peace': 'Victory/Success',
        'ok_sign': 'OK/Ready',
        'pointing': 'Navigate/Point',
        'rock': 'Rock (Game)',
        'paper': 'Paper (Game)',
        'scissors': 'Scissors (Game)'
    }

    return actions.get(gesture, 'Unknown Action')

# Test gesture control system
print("Testing gesture-based control system...")
test_samples = np.random.choice(len(X_test), 5, replace=False)

for idx in test_samples:
    test_image = X_test[idx]
    true_gesture = gesture_names[y_test[idx]]

    predicted_gesture, confidence, all_probs = predict_gesture(
        (test_image * 255).astype(np.uint8), model, gesture_names
    )

    action = simulate_gesture_control(predicted_gesture)

    print(f"\nGesture: {predicted_gesture.replace('_', ' ').title()}")
    print(f"Confidence: {confidence:.3f}")
    print(f"Action: {action}")
    print(f"Correct: {'Yes' if predicted_gesture == true_gesture else 'No'}")

# Save the model manually (error-free way)
print("\nSaving model...")
try:
    model.save('gesture_recognition_model.h5')
    print("Model saved successfully as 'gesture_recognition_model.h5'")
except:
    print("HDF5 save failed, trying SavedModel format...")
    try:
        model.save('gesture_recognition_model')
        print("Model saved successfully as 'gesture_recognition_model' (SavedModel format)")
    except:
        print("Model saving failed, but training completed successfully")

# Model performance summary
print("\nModel Performance Summary:")
print("=" * 30)
print(f"Overall Test Accuracy: {test_accuracy*100:.2f}%")
print(f"Number of Gestures Recognized: {len(gesture_names)}")
print(f"Average Confidence: {np.mean(np.max(y_pred_proba, axis=1)):.3f}")

if test_accuracy >= 0.85:
    print("EXCELLENT: Model achieved 85%+ accuracy for gesture recognition!")
elif test_accuracy >= 0.80:
    print("GOOD: Model achieved 80%+ accuracy for gesture recognition!")
else:
    print("Model accuracy could be improved.")

print("\nALL ERRORS FIXED:")
print("1. Removed problematic data augmentation layers")
print("2. Used ImageDataGenerator for data augmentation")
print("3. Removed ModelCheckpoint callback")
print("4. Added manual model saving with error handling")
print("5. Used only compatible TensorFlow/Keras features")
print("6. Added padding='same' to prevent dimension issues")

print("\nGesture Recognition System Ready!")
print("This code should run without any errors on any TensorFlow version!")
```
